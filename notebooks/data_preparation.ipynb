```python
def load_data(fin):
    '''accepts fin a string representing file name
    returns list of lines in file'''
    X = []
    y = []
    with open(fin, encoding="utf8") as f:
    
        
        labels = next(f).rstrip('\n').split('\t')
        for line in f:
            parts = line.rstrip('\n').split('\t')
            if len(set(parts)) != 3:
                continue
            for part, label in zip(parts, labels):
                X.append(part)
                y.append(label)
    return X,y
```

```python
X,y = load_data('F:/classification/data/z_l_d-sents.tsv')
```

```python
# inpsect top k instances
k = 80
for i,j in zip(y[:k], X[:k]):
    print(i,j)
```

```python
# generating character ngrams
# 2 and 3 ngrams --> window sizes 

ngram_tokens = defaultdict(int)
for n in (2, 3):
    for x in X:
        for ngram in ngrams(x, n):
            if ' ' in ngram or '።' in ngram:
                continue
            ngram = ''.join(ngram)
            ngram_tokens[ngram] += 1
```

```python
def gen_ngrams(x, ns=(2, 3)):
    for n in ns:
        for ngram in ngrams(x,n):
            if ' ' in ngram or '።' in ngram:
                continue
            ngram = ''.join(ngram)
            yield ngram
```

```python
#TODO hyper parameter to be optimized
#take top 128 | 200 | 300 --> n-vector-size or vocabulary size 

top_ngrams_freq = sorted(ngram_tokens.items(), key=lambda x:x[1], reverse=True)
ngrams_128 = [ngram for ngram, freq in top_ngrams_freq][:128]
ngrams_200 = [ngram for ngram, freq in top_ngrams_freq][:200]
ngrams_300 = [ngram for ngram, freq in top_ngrams_freq][:300]
ngrams_400 = [ngram for ngram, freq in top_ngrams_freq][:400]
ngrams_500 = [ngram for ngram, freq in top_ngrams_freq][:500]
```

```python
def get_ngram_features(X, k_ngrams):
    _X = []
    for x in X:
        x_ngrams = {ngram for ngram in gen_ngrams(x)}
        feat = [1 if ngram in x_ngrams else 0 for ngram in k_ngrams]
        _X.append(feat)
    return _X
```

```python
# save features to disk
def save_features(y, X, fname):
    with open(fname, 'w') as f:
        for _y, x in zip(y, X):
            x = ' '.join([str(i) for i in x])
            f.write(f'{_y}\t{x}\n')
```

```python
# number of ngrams
_X = get_ngram_features(X, ngrams_128)
save_features(y, _X, 'F:/classification/data/ngram_128.tsv')

_X = get_ngram_features(X, ngrams_200)
save_features(y, _X, 'F:/classification/data/ngram_200.tsv')

_X = get_ngram_features(X, ngrams_300)
save_features(y, _X, 'F:/classification/data/ngram_300.tsv')

_X = get_ngram_features(X, ngrams_400)
save_features(y, _X, 'F:/classification/data/ngram_400.tsv')

_X = get_ngram_features(X, ngrams_500)
save_features(y, _X, 'F:/classification/data/ngram_500.tsv')
```
